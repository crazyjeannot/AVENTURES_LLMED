```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_palette('bright')
sns.set_context("talk")
sns.set_style("darkgrid")
```

```python
df_main = pd.read_csv('../data/preds_Education.csv')
```

```python
df_main.Preds.value_counts()
```

```python
test = df_main[df_main['Proba_A']>0.8]
```

```python
test.index = test['Index']
```

```python
df_main['chunk_number'] = df_main["Index"].apply(lambda x: int(x.split('_')[-1]))
```

```python
df_main
```

```python
plt.figure(figsize=(9,6))
ax = plt.subplot()

ax = sns.regplot(x='chunk_number', y='Proba_A', data=df_main, order=5)

ax.set_xlabel("Progression of the narrative")
ax.set_ylabel('Probability for Adventure Scene')

plt.savefig('../img/education_sentimentale.png', format='png', bbox_inches='tight', dpi=400)

plt.show()
```
